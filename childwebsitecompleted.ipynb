```python
import pandas as pd

# Not limiting the column number when displaying dataframe
pd.set_option("display.max_columns", None)
```

```python
from sklearn import preprocessing
df=pd.DataFrame(pd.read_excel("children.xlsx"))
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.Sex=le.fit_transform(df['Sex'])
df.Ethnicity=le.fit_transform(df['Ethnicity'])
```

```python
df.Family_mem_with_ASD=le.fit_transform(df['Family_mem_with_ASD'])
```

```python
df.Jaundice=le.fit_transform(df['Jaundice'])
df = df.replace("?", 0)
df = df.replace("YES", '   ASD Trait')
df = df.replace("NO", '   Not ASD Trait')
```

```python
df.head(18)
```

```python
x=df[['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','Age_Mons','Sex','Jaundice']].values
```

```python
y=df['ASD'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.35)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = "gini")
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver = "liblinear")
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

```python
clf= LogisticRegression()
clf.fit(xtest,ytest)
```

```python
import joblib
joblib.dump(clf,"clf.pkl")
```

```python
%%writefile app.py
import pandas as pd

import streamlit as st
import pandas as pd
import joblib

# Title
st.header("Autism Detection App for Children")


A1=st.selectbox('Q1. S/he often notices small sounds when others do not',['yes','no'])

A2=st.selectbox('Q2. S/he usually concentrates more on the whole picture, rather than the small details',['yes','no'])

A3=st.selectbox('Q3. In a social group, s/he can easily keep track of several different people’s conversations',['yes','no'])

A4=st.selectbox('Q4. S/he finds it easy to go back and forth between different activities',['yes','no'])

A5=st.selectbox('Q5. S/he doesn’t know how to keep a conversation going with his/her peers',['yes','no'])

A6=st.selectbox('Q6. S/he is good at social chit-chat',['yes','no'])

A7=st.selectbox('Q7. When s/he is read a story, s/he finds it difficult to work out the character’s intentions or feelings',['yes','no'])

A8=st.selectbox('Q8. When s/he was in preschool, s/he used to enjoy playing games involving pretending with other children',['yes','no'])

A9=st.selectbox('Q9.S/he finds it easy to work out what someone is thinking or feeling just by looking at their face',['yes','no'])

A10=st.selectbox('Q10. S/he finds it hard to make new friend',['yes','no'])

# Input bar 1
Age_Mons = st.number_input("Enter Age", min_value=1, max_value=18)

# Input bar 2
#weight = st.number_input("Enter Weight")

# Dropdown input
#eyes = st.selectbox("Select Eye Colour", ("Blue", "Brown"))
Sex = st.selectbox("Select Sex", ("m", "f"))
Jaundice = st.selectbox("Select previous Jaundice", ("yes", "no"))

# If button is pressed
#if st.button("Submit")
if st.button('submit'):
  clf=joblib.load('clf.pkl')
  xtest= pd.DataFrame([[A1,A2,A3,A4,A5,A6,A7,A8,A9,A10,Age_Mons,Sex,Jaundice]], columns=['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','Age_Mons','Sex','Jaundice'])
  xtest=xtest.replace(['yes','no'],[1,0])
  xtest=xtest.replace(['m','f'],[1,0])
  prediction=clf.predict(xtest)[0]
  st.text(f"this is{prediction}")
```

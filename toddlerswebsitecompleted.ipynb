```python
import pandas as pd

# Not limiting the column number when displaying dataframe
pd.set_option("display.max_columns", None)
```

```python
from sklearn import preprocessing
df=pd.DataFrame(pd.read_excel("toddlers.xlsx"))
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.Sex=le.fit_transform(df['Sex'])
df.Ethnicity=le.fit_transform(df['Ethnicity'])
```

```python
df.Family_mem_with_ASD=le.fit_transform(df['Family_mem_with_ASD'])
```

```python
df.Jaundice=le.fit_transform(df['Jaundice'])
df = df.replace("?", 0)
df = df.replace("Yes", '   ASD Trait')
df = df.replace("No", '   Not ASD Trait')
```

```python
df.head(18)
```

```python
x=df[['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','Age_Mons','Sex','Jaundice']].values
```

```python
y=df['ASD'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.35)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = "gini")
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver = "liblinear")
model.fit(xtrain, ytrain)

print(model.score(xtest, ytest))
```

```python
clf= LogisticRegression()
clf.fit(xtest,ytest)
```

```python
import joblib
joblib.dump(clf,"clf.pkl")
```

```python
%%writefile app.py
import pandas as pd

import streamlit as st
import pandas as pd
import joblib

# Title
st.header("Autism Detection App for Toddlers")


A1=st.selectbox('Q1. Does your child look at you when you call his/her name?',['yes','no'])

A2=st.selectbox('Q2. How easy is it for you to get eye contact with your child?',['yes','no'])

A3=st.selectbox('Q3. Does your child point to indicate that s/he wants something? (e.g., a toy that is out of reach)',['yes','no'])

A4=st.selectbox('Q4. Does your child point to share interest with you? (e.g., pointing at aninteresting sight)',['yes','no'])

A5=st.selectbox('Q5. S/he doesn’t know how to keep a conversation going with his/her peersDoes your child pretend? (e.g., care for dolls, talk on a toy phone)',['yes','no'])

A6=st.selectbox('Q6.Does your child follow where you’re looking?',['yes','no'])

A7=st.selectbox('Q7. If you or someone else in the family is visibly upset, does your child show signsof wanting to comfort them? (e.g., stroking hair, hugging them)',['yes','no'])

A8=st.selectbox('Q8. Would you describe your child’s first words as:',['yes','no'])

A9=st.selectbox('Q9.Does your child use simple gestures? (e.g., wave goodbye)',['yes','no'])

A10=st.selectbox('Q10. Does your child stare at nothing with no apparent purpose?',['yes','no'])

# Input bar 1
Age_Mons = st.number_input("Enter Age in months", min_value=12, max_value=36)

# Input bar 2
#weight = st.number_input("Enter Weight")

# Dropdown input
#eyes = st.selectbox("Select Eye Colour", ("Blue", "Brown"))
Sex = st.selectbox("Select Sex", ("m", "f"))
Jaundice = st.selectbox("Select previous Jaundice", ("yes", "no"))

# If button is pressed
#if st.button("Submit")
if st.button('submit'):
  clf=joblib.load('clf.pkl')
  xtest= pd.DataFrame([[A1,A2,A3,A4,A5,A6,A7,A8,A9,A10,Age_Mons,Sex,Jaundice]], columns=['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','Age_Mons','Sex','Jaundice'])
  xtest=xtest.replace(['yes','no'],[1,0])
  xtest=xtest.replace(['m','f'],[1,0])
  prediction=clf.predict(xtest)[0]
  st.text(f"this is{prediction}")
```
